```python
# Prepare the text data and timestamps
texts = df['comments'].tolist()  #  reviews
timestamps = pd.to_datetime(df['date']).tolist()  #  timestamps

```

```python
from bertopic import BERTopic
```

## Visualize Topics over Time

```python
# Initialize BERTopic model
topic_model = BERTopic(verbose=True)

# Fit the model to documents
topics, _ = topic_model.fit_transform(texts)
```

```python
topics_over_time = topic_model.topics_over_time(texts, timestamps)
topic_model.visualize_topics_over_time(topics_over_time, top_n_topics=20)
```

```python
# visualize selected topic

selected_topic = [2, 6, 13]

topic_model.visualize_topics_over_time(topics_over_time, topics=selected_topic)
```

The data shows that topic 2, which seems related to positive experiences, rose in Q2 and declined in Q3. However, the increase and subsequent decrease may be explained by seasonal trends rather than being solely attributable to the new law enacted. The peak in Q2 and decline in Q3 could reflect a summer holiday peak effect. More data over a longer period of time would be needed to fully understand the factors driving the changes observed in topic 2.
